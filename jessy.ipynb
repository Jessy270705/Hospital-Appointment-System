```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

plt.figure(figsize=(8,5))

sns.countplot(
    x='Gender',
    data=df,
    palette={'Male': 'blue', 'Female': 'purple'}
)

plt.title("Gender Count", fontsize=14, fontweight='bold')
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
df['School_Type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("School Type Distribution")
plt.ylabel("")
plt.show()
```

```python


sns.set_style("whitegrid")

plt.figure(figsize=(8,5))

sns.boxplot(
    x='Gender',
    y='Exam_Score',
    data=df,
    palette={'Male': '#1f77b4', 'Female': '#8e44ad'}  # Blue & Purple (professional shades)
)

plt.title("Gender vs Exam Score", fontsize=14, fontweight='bold')
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Exam Score", fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Teacher_Quality',
    y='Exam_Score',
    data=df,
    palette='viridis'
)

plt.title("Teacher Quality vs Exam Score", fontsize=14, fontweight='bold')
plt.show()
```

```python
hrs_by_score=df.groupby('Hours_Studied')['Exam_Score'].sum()
top10=hrs_by_score.head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=top10.index,y=top10.values,hue=top10.index,palette="viridis")
plt.title("First 10 Studied Hours by Exam Score")
plt.xlabel("Hours_Studied")
plt.ylabel("Exam_Score")
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```

# Insights From Above Charts
# .There is a positive relationship between study hours and exam scores
# .Students with good past performance continue to perform well
# .Higher motivation levels correspond to higher average scores
# .Better teacher quality leads to improved student performance
# .No major difference in median scores between genders
